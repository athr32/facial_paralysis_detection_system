```python
from ultralytics import YOLO

# Step 1: Train the YOLOv8 model
model = YOLO("yolov8m.pt")  # Start with a pre-trained YOLOv8 model

# Train configuration
results = model.train(
    data="/content/drive/MyDrive/facial_paralysis/data.yaml",  # Path to your data.yaml
    epochs=150,
    patience=0,# Number of epochs
    imgsz=640,                      # Image size for training
    batch=16,
    augment=True ,
    # Batch size (adjust based on GPU memory)
    name="track_symbol_detection"   # Project name
)

# Save the best trained model
print("Model training complete.")
```

```python
import cv2
from datetime import datetime
from ultralytics import YOLO
from google.colab.patches import cv2_imshow
from google.colab import files
import random

# Load the trained YOLOv8 model
model = YOLO("/content/runs/detect/track_symbol_detection/weights/best.pt")  # Update path if needed

# (Optional) If needed: Force manually setting class names
# model.names = {0: 'Normal', 1: 'Paralysis'}

# Get patient name
patient_name = input("Enter the patient's name: ")

# Generate unique colors for each class
class_colors = {class_name: (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))
                for class_name in model.names.values()}

# Upload video or images
uploaded = files.upload()
file_name = list(uploaded.keys())[0]

# Check if uploaded file is a video or an image
if file_name.endswith((".mp4", ".avi", ".mov")):  # Video
    cap = cv2.VideoCapture(file_name)

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Detection
        results = model(frame)
        detections = results[0].boxes.data.cpu().numpy()

        for detection in detections:
            x1, y1, x2, y2, confidence, class_id = detection
            class_name = model.names[int(class_id)]

            # Debug: See the detected class
            print(f"Detected Class: {class_name}")

            # Determine result
            is_having_paralysis = "Yes" if class_name.lower() == "paralysis" else "No"

            # Drawing
            color = class_colors[class_name]
            confidence = round(float(confidence) * 100, 2)
            label = f"{class_name}: {confidence}%"
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), color, 2)
            cv2.putText(frame, label, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

            # Print results
            timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            print(f"Patient: {patient_name} | Time: {timestamp} | Paralysis Detected: {is_having_paralysis}")

        cv2_imshow(frame)

    cap.release()

elif file_name.endswith((".jpg", ".png", ".jpeg")):  # Image
    frame = cv2.imread(file_name)

    # Detection
    results = model(frame)
    detections = results[0].boxes.data.cpu().numpy()

    for detection in detections:
        x1, y1, x2, y2, confidence, class_id = detection
        class_name = model.names[int(class_id)]

        # Debug: See the detected class
        print(f"Detected Class: {class_name}")

        # Determine result
        is_having_paralysis = "Yes" if class_name.lower() == "paralysis" else "No"

        # Drawing
        color = class_colors[class_name]
        confidence = round(float(confidence) * 100, 2)
        label = f"{class_name}: {confidence}%"
        cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), color, 2)
        cv2.putText(frame, label, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

        # Print results
        timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        print(f"Patient: {patient_name} | Time: {timestamp} | Paralysis Detected: {is_having_paralysis}")

    cv2_imshow(frame)

```
